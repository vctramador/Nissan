```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# carregando dados
df = pd.read_csv('nissan-dataset.csv')
print(df.info())
# verificando valores nulos
print(df.isnull().sum())
# descrevendo estatisticas
print(df.describe())
```

```python
df
```

```python
#deletando colunas com possiveis valores nulos
df = df.dropna()
print(df.isnull().sum())
```

```python
# estatisticas resumidas

# PreÃ§o
print('Estatisticas de preÃ§o')
print(f"Menor preÃ§o: {df['price'].min()}")
print(f"Maior preÃ§o: {df['price'].max()}")
print(f"Media de preÃ§o: {df['price'].mean()}")

# idade
print('Estatisticas de idade')
print(f"Menor idade: {df['age'].min()}")
print(f"Maior idade: {df['age'].max()}")
print(f"Media de idade: {df['age'].mean()}")

# kilometros
print('Estatisticas de kilometros')
print(f"Baixa kilometragem: {df['km'].min()}")
print(f"Alta kilometragem: {df['km'].max()}")
print(f"Media de kilometragem: {df['km'].mean()}")
```

```python
# estatisticas de consumidores

# Histograma de distribuiÃ§Ã£o etÃ¡ria
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('DistribuiÃ§Ã£o de Idades dos ProprietÃ¡rios')
plt.xlabel('Idade')
plt.ylabel('FrequÃªncia')
plt.savefig('Distribuicao_de_Idade.png')
plt.show()

# GrÃ¡fico de barras da distribuiÃ§Ã£o por gÃªnero
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('DistribuiÃ§Ã£o por GÃªnero dos ProprietÃ¡rios')
plt.xlabel('GÃªnero')
plt.ylabel('FrequÃªncia')
plt.savefig('Distribuicao_genero.png')
plt.xticks(rotation=-90) # rotacionando o eixo x para melhor visualizaÃ§Ã£o
plt.show()

# Idade mÃ©dia por gÃªnero
idade_x_genero = df.groupby('gender')['age'].mean().reset_index()
print(idade_x_genero)

# GrÃ¡fico de barras da idade mÃ©dia por gÃªnero
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='age', data=idade_x_genero)
plt.title('Idade MÃ©dia por GÃªnero')
plt.xlabel('GÃªnero')
plt.ylabel('Idade MÃ©dia')
plt.savefig('idadexgenero.png')
plt.xticks(rotation=-90) # rotacionando o eixo x para melhor visualizaÃ§Ã£o
plt.show()
```

```python
# estatisticas de marketing

# contagem de veiculos por condiÃ§Ã£o
plt.figure(figsize=(10, 6))
sns.countplot(x='condition', data=df)
plt.title('Contagem de veiculos por condiÃ§Ã£o')
plt.xlabel('CondiÃ§Ã£o')
plt.ylabel('Contagem')
plt.savefig('contagem_condicao.png')
plt.show()

# media de preÃ§o por condiÃ§Ã£o
plt.figure(figsize=(10, 6))
sns.barplot(x='condition', y='price', data=df, palette='viridis')
plt.title('MÃ©dia de preÃ§o por condiÃ§Ã£o')
plt.xlabel('CondiÃ§Ã£o')
plt.ylabel('MÃ©dia de preÃ§o')
plt.savefig('precoxcondicao.png')
plt.show()

# DistribuiÃ§Ã£o de idade por gÃªnero
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='age', data=df)
plt.title('DistribuiÃ§Ã£o de Idade por GÃªnero')
plt.xlabel('GÃªnero')
plt.ylabel('Idade')
plt.savefig('idadexcondicao.png')
plt.show()

# RelaÃ§Ã£o entre preÃ§o e atributos (idade, performance, quilometragem)
plt.figure(figsize=(10, 6))
sns.heatmap(df[['age', 'performance', 'km', 'price']].corr(), annot=True, cmap='coolwarm')
plt.title('CorrelaÃ§Ã£o entre Atributos e PreÃ§o')
plt.show()

# Impacto da condiÃ§Ã£o no preÃ§o
plt.figure(figsize=(10, 6))
sns.boxplot(x='condition', y='price', data=df)
plt.title('Impacto da CondiÃ§Ã£o no PreÃ§o')
plt.xlabel('CondiÃ§Ã£o')
plt.ylabel('PreÃ§o ($)')
plt.show()


```
