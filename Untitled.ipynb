```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# carregando dados
df = pd.read_csv('nissan-dataset.csv')
print(df.info())
# verificando valores nulos
print(df.isnull().sum())
# descrevendo estatisticas
print(df.describe())
```

```python
df
```

```python
#deletando colunas com possiveis valores nulos
df = df.dropna()
print(df.isnull().sum())
```

```python
# estatisticas resumidas

# Preço
print('Estatisticas de preço')
print(f"Menor preço: {df['price'].min()}")
print(f"Maior preço: {df['price'].max()}")
print(f"Media de preço: {df['price'].mean()}")

# idade
print('Estatisticas de idade')
print(f"Menor idade: {df['age'].min()}")
print(f"Maior idade: {df['age'].max()}")
print(f"Media de idade: {df['age'].mean()}")

# kilometros
print('Estatisticas de kilometros')
print(f"Baixa kilometragem: {df['km'].min()}")
print(f"Alta kilometragem: {df['km'].max()}")
print(f"Media de kilometragem: {df['km'].mean()}")
```

```python
# estatisticas de consumidores

# Histograma de distribuição etária
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribuição de Idades dos Proprietários')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.savefig('Distribuicao_de_Idade.png')
plt.show()

# Gráfico de barras da distribuição por gênero
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Distribuição por Gênero dos Proprietários')
plt.xlabel('Gênero')
plt.ylabel('Frequência')
plt.savefig('Distribuicao_genero.png')
plt.xticks(rotation=-90) # rotacionando o eixo x para melhor visualização
plt.show()

# Idade média por gênero
idade_x_genero = df.groupby('gender')['age'].mean().reset_index()
print(idade_x_genero)

# Gráfico de barras da idade média por gênero
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='age', data=idade_x_genero)
plt.title('Idade Média por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Idade Média')
plt.savefig('idadexgenero.png')
plt.xticks(rotation=-90) # rotacionando o eixo x para melhor visualização
plt.show()
```

```python
# estatisticas de marketing

# contagem de veiculos por condição
plt.figure(figsize=(10, 6))
sns.countplot(x='condition', data=df)
plt.title('Contagem de veiculos por condição')
plt.xlabel('Condição')
plt.ylabel('Contagem')
plt.savefig('contagem_condicao.png')
plt.show()

# media de preço por condição
plt.figure(figsize=(10, 6))
sns.barplot(x='condition', y='price', data=df, palette='viridis')
plt.title('Média de preço por condição')
plt.xlabel('Condição')
plt.ylabel('Média de preço')
plt.savefig('precoxcondicao.png')
plt.show()

# Distribuição de idade por gênero
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='age', data=df)
plt.title('Distribuição de Idade por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Idade')
plt.savefig('idadexcondicao.png')
plt.show()

# Relação entre preço e atributos (idade, performance, quilometragem)
plt.figure(figsize=(10, 6))
sns.heatmap(df[['age', 'performance', 'km', 'price']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlação entre Atributos e Preço')
plt.show()

# Impacto da condição no preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='condition', y='price', data=df)
plt.title('Impacto da Condição no Preço')
plt.xlabel('Condição')
plt.ylabel('Preço ($)')
plt.show()


```
